```python
try:
    dbConnection = psycopg2.connect(
    user = "postgres",
    password = "password",
    #host = "192.168.56.30",
    host = "localhost",
    port = "5432",
    database = "covid")
    dbConnection.set_isolation_level(0) # AUTOCOMMIT
    dbCursor = dbConnection.cursor()
    insertString = "INSERT INTO exr VALUES('{}','{}','{}')"
    for i in range(len(df)):
        Date =df.iloc[i:i+1,:]['Date'].values[0]
        Currency =df.iloc[i:i+1,:]['Currency'].values[0]
        Rate =df.iloc[i:i+1,:]['Rate'].values[0]
        row = Date, Currency, Rate
        dbCursor.execute(insertString.format(*row))
    dbConnection.commit()
    dbCursor.close()
except (Exception , psycopg2.Error) as dbError :
    print ("Error:", dbError)
    raise
finally:
    if(dbConnection): dbConnection.close() 
```
